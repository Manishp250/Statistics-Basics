# Statistics Basics

### 1. What is statistics, and why is it important?
Statistics is the science of collecting, analyzing, interpreting, presenting, and organizing data. It's important because it helps us make informed decisions based on data, identify patterns, test hypotheses, and draw reliable conclusions from observations.

### 2. What are the two main types of statistics?
The two main types are:
- Descriptive statistics: Summarize and describe data
- Inferential statistics: Make predictions or inferences about a population based on sample data

### 3. What are descriptive statistics?
Descriptive statistics are methods used to summarize and describe the main features of a dataset, including measures of central tendency (mean, median, mode) and measures of variability (range, variance, standard deviation).

### 4. What is inferential statistics?
Inferential statistics uses sample data to make generalizations about a larger population, including hypothesis testing, confidence intervals, and regression analysis.

### 5. What is sampling in statistics?
Sampling is the process of selecting a subset of individuals from a population to estimate characteristics of the whole population.

### 6. What are the different types of sampling methods?
Common sampling methods include:
- Random sampling
- Stratified sampling
- Systematic sampling
- Cluster sampling
- Convenience sampling

### 7. What is the difference between random and non-random sampling?
Random sampling gives each member of the population an equal chance of being selected, while non-random sampling doesn't use random selection (e.g., convenience sampling).

### 8. Define and give examples of qualitative and quantitative data.
- Qualitative: Non-numerical data (e.g., colors, brands, categories)
- Quantitative: Numerical data (e.g., height, weight, temperature)

### 9. What are the different types of data in statistics?
Main types:
- Nominal (categories without order)
- Ordinal (ordered categories)
- Interval (ordered with equal intervals)
- Ratio (interval with true zero)

### 10. Explain nominal, ordinal, interval, and ratio levels of measurement.
- Nominal: Categories without order (e.g., gender, color)
- Ordinal: Ordered categories (e.g., satisfaction levels)
- Interval: Equal intervals, no true zero (e.g., temperature in Â°C)
- Ratio: Equal intervals with true zero (e.g., height, weight)

### 11. What is the measure of central tendency?
A measure of central tendency is a single value that represents the center point of a dataset (mean, median, or mode).

### 12. Define mean, median, and mode.
- Mean: The average of all values
- Median: The middle value when data is ordered
- Mode: The most frequent value

### 13. What is the significance of the measure of central tendency?
It provides a single value that represents the entire dataset, making it easier to understand and compare different datasets.

### 14. What is variance, and how is it calculated?
Variance measures how far each number in the set is from the mean. It's calculated as the average of the squared differences from the mean.

### 15. What is standard deviation, and why is it important?
Standard deviation is the square root of variance. It's important because it measures dispersion in the same units as the original data.

### 16. Define and explain the term range in statistics.
Range is the difference between the highest and lowest values in a dataset, providing a simple measure of variability.

### 17. What is the difference between variance and standard deviation?
Variance is the average squared deviation from the mean, while standard deviation is the square root of variance, expressed in the original units.

### 18. What is skewness in a dataset?
Skewness measures the asymmetry of the probability distribution of a real-valued random variable.

### 19. What does it mean if a dataset is positively or negatively skewed?
- Positively skewed: The tail on the right side is longer
- Negatively skewed: The tail on the left side is longer

### 20. Define and explain kurtosis.
Kurtosis measures the "tailedness" of the probability distribution, indicating whether data are heavy-tailed or light-tailed relative to a normal distribution.

### 21. What is the purpose of covariance?
Covariance measures how much two random variables change together.

### 22. What does correlation measure in statistics?
Correlation measures the strength and direction of the linear relationship between two variables.

### 23. What is the difference between covariance and correlation?
Covariance indicates direction but not strength, while correlation standardizes covariance to a range of -1 to 1.

### 24. What are some real-world applications of statistics?
- Medical research
- Market analysis
- Quality control
- Weather forecasting
- Sports analytics
- Political polling

```python
# 10. Write a python script to calculate the correlation coefficient between two datasets.

import numpy as np

def calculate_correlation(x, y):
    """
    Calculate the correlation coefficient between two datasets.

    Parameters:
    x (list or np.array): First dataset
    y (list or np.array): Second dataset

    Returns:
    float: Correlation coefficient
    """
    # Convert to numpy arrays
    x = np.array(x)
    y = np.array(y)

    # Ensure both datasets have the same length
    if len(x) != len(y):
        raise ValueError("Datasets must have the same length.")

    # Calculate the correlation coefficient
    correlation = np.corrcoef(x, y)[0, 1]

    return correlation

# Example datasets
data1 = [10, 20, 30, 40, 50]
data2 = [15, 25, 35, 45, 55]

# Calculate correlation
result = calculate_correlation(data1, data2)
print(f"The correlation coefficient is: {result}")
```

```python
# 12. Implement and compare simple random sampling and systematic sampling.

import numpy as np

# Simple random sampling
def simple_random_sample(data, n):
    return np.random.choice(data, size=n, replace=False)

# Systematic sampling
def systematic_sample(data, n):
    k = len(data) // n
    indices = np.arange(0, len(data), k)
    return data[indices]

# Example usage
data = np.arange(1, 101)
simple_sample = simple_random_sample(data, 50)
systematic_sample = systematic_sample(data, 50)
print(f"Simple Random Sample: {simple_sample}")
print(f"Systematic Sample: {systematic_sample}")
```

```python
# 13. Calculate the mean, median, and mode of grouped data.

import numpy as np
from collections import Counter

def calculate_mean(frequencies, midpoints):
    total_freq = sum(frequencies)
    sum_fx = sum(f * x for f, x in zip(frequencies, midpoints))
    mean = sum_fx / total_freq
    return mean

def calculate_median(class_intervals, frequencies):
    cumulative_frequencies = np.cumsum(frequencies)
    total_freq = cumulative_frequencies[-1]
    N_2 = total_freq / 2

    # Find the median class
    for i, cum_freq in enumerate(cumulative_frequencies):
        if cum_freq >= N_2:
            median_class_index = i
            break

    L = class_intervals[median_class_index][0]
    fm = frequencies[median_class_index]
    cf = cumulative_frequencies[median_class_index - 1] if median_class_index != 0 else 0
    h = class_intervals[0][1] - class_intervals[0][0]

    median = L + ((N_2 - cf) / fm) * h
    return median

def calculate_mode(class_intervals, frequencies):
    modal_class_index = frequencies.index(max(frequencies))

    L = class_intervals[modal_class_index][0]
    f1 = frequencies[modal_class_index]
    f0 = frequencies[modal_class_index - 1] if modal_class_index != 0 else 0
    f2 = frequencies[modal_class_index + 1] if modal_class_index < len(frequencies) - 1 else 0
    h = class_intervals[0][1] - class_intervals[0][0]

    mode = L + ((f1 - f0) / ((f1 - f0) + (f1 - f2))) * h
    return mode

# Example Data
class_intervals = [(10, 20), (20, 30), (30, 40), (40, 50)]
frequencies = [5, 8, 12, 6]

# Calculate midpoints
midpoints = [(interval[0] + interval[1]) / 2 for interval in class_intervals]

# Calculations
mean = calculate_mean(frequencies, midpoints)
median = calculate_median(class_intervals, frequencies)
mode = calculate_mode(class_intervals, frequencies)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
```

```python
# 18. Implement Z-score normalization and explain its significance.

import numpy as np

def z_score_normalization(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    normalized_data = (data - mean) / std_dev
    return normalized_data

# Example data
data = np.random.normal(loc=50, scale=15, size=100)

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Perform Z-score normalization
z_scores = (data - mean) / std_dev

print(f"Original Data: {data[:5]}")  # Show first 5 values for reference
print(f"Z-scores: {z_scores[:5]}")   # Show first 5 Z-scores for reference

# Significance of Z-score Normalization:

# 1. Comparing Data on Different Scales: It helps when dealing with data having different units or scales (e.g., height in cm, weight in kg) by standardizing them to a common scale.


# 2. Improves Convergence in Machine Learning: In algorithms like gradient descent, Z-score normalization can help the optimization process converge faster, as features are on the same scale.


# 3. Outlier Detection: Z-scores can be used to identify outliers by checking for values with a Z-score above or below a certain threshold (usually 2 or -2).
```

```python
# 24. Implement the Pearson and Spearman correlation coefficients for a dataset.

# Pearson Correlation: Measures the linear relationship between two variables. It ranges from -1 to 1, where 1 indicates a perfect positive linear relationship, -1 indicates a perfect negative linear relationship, and 0 indicates no linear relationship.

# Spearman Correlation: Measures the monotonic relationship between two variables. It is based on the ranks of the data rather than the actual values, making it more robust to outliers.

import numpy as np
from scipy.stats import pearsonr, spearmanr

# Generate random datasets
np.random.seed(42)  # Ensures reproducibility
data1 = np.random.randint(1, 100, 100)
data2 = np.random.randint(1, 100, 100)

# Calculate Pearson Correlation
pearson_corr, pearson_p_value = pearsonr(data1, data2)

# Calculate Spearman Correlation
spearman_corr, spearman_p_value = spearmanr(data1, data2)

# Display the results
print(f"Pearson Correlation: {pearson_corr:.2f}, P-value: {pearson_p_value:.4f}")
print(f"Spearman Correlation: {spearman_corr:.2f}, P-value: {spearman_p_value:.4f}")

```
